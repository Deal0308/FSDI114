```python
#  Simploed implentations of Queue of Queue (using built_in data Structue)

class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, value):
        self.items.insert(0, value)
        
    def dequeue(self):
        return self.items.pop()


    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []
        
```

```python
#  From scratch implementation of Queue

class QueueII:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.front = None
        self.rear = None
        self.__size = 0  # Rename the attribute to __size

    def enqueue(self, value):
        new_node = self.__Node(value)
        if self.is_empty():
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
        self.__size += 1  # Update the attribute

    def dequeue(self):
        if self.is_empty():
            raise IndexError("Queue is empty")
        value = self.front.data
        self.front = self.front.next
        self.__size -= 1  # Update the attribute
        if self.is_empty():
            self.rear = None
        return value

    def peek(self):
        if self.is_empty():
            raise IndexError("Queue is empty")
        return self.front.data

    def get_size(self):  # Rename the method to get_size
        return self.__size  # Update the method name

    def is_empty(self):
        return self.__size == 0  # Update the attribute name

    def __str__(self):
        current = self.front
        result = []
        while current:
            result.append(str(current.data))
            current = current.next
            if current:
                result.append("->")
        return "".join(result)


queue = QueueII()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)

print(str(queue))      
print(queue.get_size())  
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self, /, *args, **kwargs):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def insert(self, index, data):
        if index < 0:
            raise IndexError("Index out of range")
        new_node = Node(data)
        if index == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            for _ in range(index - 1):
                if not current:
                    raise IndexError("Index out of range")
                current = current.next
            new_node.next = current.next
            current.next = new_node

    def remove(self, data):
        current = self.head
        if current and current.data == data:
            self.head = current.next
            return
        while current and current.next:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next
        raise ValueError(f"{data} not found in the list")

    def index(self, data):
        current = self.head
        index = 0
        while current:
            if current.data == data:
                return index
            current = current.next
            index += 1
        raise ValueError(f"{data} not found in the list")

    def copy(self):
        new_list = SinglyLinkedList()
        current = self.head
        while current:
            new_list.append(current.data)
            current = current.next
        return new_list

    def delete_all(self, data):
        current = self.head
        while current and current.data == data:
            self.head = current.next
            current = self.head
        while current and current.next:
            if current.next.data == data:
                current.next = current.next.next
            else:
                current = current.next

    def __getitem__(self, index):
        current = self.head
        for _ in range(index):
            if not current:
                raise IndexError("Index out of range")
            current = current.next
        if not current:
            raise IndexError("Index out of range")
        return current.data

    def __str__(self):
        result = []
        current = self.head
        while current:
            result.append(str(current.data))
            current = current.next
            if current:
                result.append(" -> ")
        return "".join(result)

# Example usage
sll = SinglyLinkedList()
sll.append(1)
sll.append(2)
sll.append(3)

print("Original SLL:", sll)

# Insert at index 1
sll.insert(1, 5)
print("After insert at index 1:", sll)

# Remove element 2
sll.remove(2)
print("After removing 2:", sll)

# Find index of element 5
index_of_5 = sll.index(5)
print("Index of 5:", index_of_5)

# Another round
sll.insert(0, 10)
print("After insert at index 0:", sll)

sll.remove(10)
print("After removing 10:", sll)

index_of_3 = sll.index(3)
print("Index of 3:", index_of_3)

# Accessing elements using __getitem__
print("Element at index 0:", sll[0])
print("Element at index 1:", sll[1])

# Copy the linked list and print
copied_sll = sll.copy()
print("Copied SLL:", copied_sll)

# Delete all occurrences of 3 and print
sll.delete_all(3)
print("After deleting all occurrences of 3:", sll)
```
